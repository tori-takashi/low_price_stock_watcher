```python
import pandas as pd
stocks_df = pd.read_csv('低位株.csv')
```

```python
filtered_pbr = stocks_df[stocks_df['PBR'] >= 1.0]
print(len(filtered_pbr.index))
filtered_pbr
```

```python
filtered_equity_ratio = filtered_pbr[filtered_pbr['自己資本比率'] > 0.4]
print(len(filtered_equity_ratio))
filtered_equity_ratio
```

```python
filtered_market_value = filtered_equity_ratio[filtered_pbr['時価総額'] < 3000000000]
print(len(filtered_market_value.index))
filtered_market_value
```

```python
filtered_stock_list = filtered_market_value.sort_values(['200日移動平均乖離率'])
```
